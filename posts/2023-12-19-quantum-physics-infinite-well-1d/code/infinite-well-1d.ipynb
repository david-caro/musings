This is a representation for the energy and the wavefunctions for a particle inside an infinite potential well

Wave function:
$$
\psi(x) = \sqrt{\frac{2}{L}}\sin{\frac{n\pi}{L}x} \hspace{0.5cm}\text{where } n=1,2,3,...
$$

Energy:
$$E_n=\frac{n^2\pi^2\hbar^2}{2mL^2}$$

```python
import math
from scipy.constants import pi, hbar
from matplotlib import pyplot as plt

## Some constants
# Length of the well
L = 100
# mass of the particle
m = 1

n_to_marker = {
    1: 'k-',
    2: 'r--',
    3: 'b-.',
    4: 'g:',
    5: 'c^',
}

def psi(x: float, n: int) -> float:
    return math.sqrt(2/L) * math.sin((n * pi)*x / L)


def energy(n: int) -> float:
    return (n * pi * hbar) **2 / (2 * m * L**2)


def plot_energy_levels(max_n: int = 4):
    plt.xlabel('$x$', fontsize=20)
    num_steps = 200
    xs = [i/num_steps for i in range(0, L*num_steps)]
    for n_sub_one in range(max_n):
        n = n_sub_one + 1
        energy_at_n = energy(n)
        ys = [energy_at_n for _ in xs]
        label = f"n={n}"
        plt.plot(xs, ys, n_to_marker[n], label=label, linewidth=3)
    plt.legend(loc=4)
    plt.title(f"First {n} energy levels")
    plt.show()


def plot_wave_functions(max_n: int = 4) -> None:
    plt.xlabel('$x$', fontsize=20)
    num_steps = 200
    xs = [i/num_steps for i in range(0, L*num_steps)]
    for n_sub_one in range(max_n):
        n = n_sub_one + 1
        ys = [psi(x=x, n=n) for x in xs]
        label = f"n={n}"
        plt.plot(xs, ys, n_to_marker[n], label=label, linewidth=3)
    plt.legend(loc=4)
    plt.title(f"Wave functions for the first {n} energy levels")
    plt.show()


def plot_probabilities(max_n: int = 4) -> None:
    plt.xlabel('$x$', fontsize=20)
    num_steps = 200
    xs = [i/num_steps for i in range(0, L*num_steps)]
    for n_sub_one in range(max_n):
        n = n_sub_one + 1
        ys = [abs(psi(x=x, n=n))**2 for x in xs]
        label = f"n={n}"
        plt.plot(xs, ys, n_to_marker[n], label=label, linewidth=3)
    plt.legend(loc=4)
    plt.title(f"Probabilities to find the particle for the first {n} energy levels")
    plt.show()
```

```python
plot_energy_levels(5)
plot_wave_functions(5)
plot_probabilities(5)
```
